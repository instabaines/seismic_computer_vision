```python
# Project Crayfish data redistribution
# Images per class -
# training - 70% (15502), validation - 15% (), test - 15% () per class
# Shuffle

import os
import shutil
import random

# types = ['BP\\', 'SH\\', 'WC\\']
types = [str(x) for x in range(0,9)]
input_directory = r'D:\sesimic\data\\'
output_directory = r'D:\sesimic\balanced_data\\'
categories = ['train', 'validation', 'test']
num_of_images = 26496
# num_of_images = 100
# spilt = ()
image_split = {
    'total': num_of_images,
    'train': int(num_of_images * 0.8),
    'val': int(num_of_images * 0.1),
    'test': int(num_of_images * 0.1)
}
print(num_of_images)


def transfer_image(in_dir, out_dir, cat, group, number):
    img_path = os.path.join(in_dir, group,)
    print('source:', img_path)
    new_dirs = os.listdir(img_path)
    random.shuffle(new_dirs)
    total_num, train_num, val_num, test_num = number['total'], number['train'], number['val'], number['test']

    print(len(new_dirs))

    for i in range(total_num):
        if i % 100 == 0:
            print('iteration number', i, new_dirs[0])
        if i < train_num:
            category = cat[0]
            source = os.path.join(in_dir, group, new_dirs[0])
            destination = os.path.join(out_dir, category, group, new_dirs[0])
            print("source:", source) # print("Destination:", destination)
            shutil.copy(source, destination)
            new_dirs.pop(0)
        elif train_num <= i < val_num + train_num:
            category = cat[1]
            source = os.path.join(in_dir, group, new_dirs[0])
            destination = os.path.join(out_dir, category, group, new_dirs[0])
            shutil.copy(source, destination)
            new_dirs.pop(0)
        elif train_num + val_num <= i < train_num + val_num + test_num:
            category = cat[2]
            source = os.path.join(in_dir, group, new_dirs[0])
            destination = os.path.join(out_dir, category, group, new_dirs[0])
            shutil.copy(source, destination)
            new_dirs.pop(0)
        else:
            print('Passed', i)
    print(f'Remaining images for {group}', len(new_dirs))


for t in types:
    print(f'\nCLASS {t}')
    transfer_image(input_directory, output_directory, categories, t, image_split)


```
